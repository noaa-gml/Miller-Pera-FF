```python
import xarray
from datetime import datetime
```

### new 2024 GHG

```python
base = xarray.open_dataset('inputs/TOTALS_flx_nc_2024_GHG/EDGAR_2024_GHG_CO2_2023_TOTALS_flx.nc')
base['history'] = f'extrapolate_EDGAR.ipynb, {datetime.now()}, "inputs/TOTALS_flx_nc_2024_GHG/EDGAR_2024_GHG_CO2_2023_TOTALS_flx.nc"'

year1 = base.copy()
year2 = base.copy()
```

```python
year1['fluxes'] = base['fluxes'] * 1.03
year2['fluxes'] = base['fluxes'] * (1.03 ** 2)

year1['fluxes'].attrs['year'] = 2024
year2['fluxes'].attrs['year'] = 2025

year1.to_netcdf('inputs/TOTALS_flx_nc_2024_GHG/EDGAR_2024_GHG_CO2_2024_TOTALS_FAKE_flx.nc')
year2.to_netcdf('inputs/TOTALS_flx_nc_2024_GHG/EDGAR_2024_GHG_CO2_2025_TOTALS_FAKE_flx.nc')
```

### old v80 GhG

```python
base = xarray.open_dataset('inputs/TOTALS_emi_nc_v80/v8.0_FT2022_GHG_CO2_2022_TOTALS_emi.nc')
base['history'] = f'extrapolate_EDGAR.ipynb, {datetime.now()}, "inputs/TOTALS_emi_nc_v80/v8.0_FT2022_GHG_CO2_2022_TOTALS_emi.nc"'

year1 = base.copy()
year2 = base.copy()
```

```python
year1['emissions'] = base['emissions'] * 1.03
year2['emissions'] = base['emissions'] * (1.03 ** 2)

year1['emissions']['year'] = 2023
year1['emissions']['year'] = 2024

year1.to_netcdf('inputs/TOTALS_emi_nc_v80/v8.0_FT2022_GHG_CO2_2023_TOTALS_emi_FAKE.nc')
year2.to_netcdf('inputs/TOTALS_emi_nc_v80/v8.0_FT2022_GHG_CO2_2024_TOTALS_emi_FAKE.nc')
```
